```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Set up Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Visit the Quotes to Scrape site 10.3.2
url = 'http://quotes.toscrape.com/'
browser.visit(url)
```

```python
# Parse the HTML 10.3.2
html = browser.html
html_soup = soup(html, 'html.parser')
```

```python
# Scrape the Title 10.3.2
title = html_soup.find('h2').text
title
```

```python
# In the above code (10.3.2)
# We used our html_soup object we created earlier and 
# chained find() to it to search for the <h2 /> tag.
# We've also extracted only the text within the HTML 
# tags by adding .text to the end of the code.
```

```python
# Scrape the top ten tags 10.3.2
tag_box = html_soup.find('div', class_='tags-box')
# tag_box
tags = tag_box.find_all('a', class_='tag')

for tag in tags:
    word = tag.text
    print(word)
```
